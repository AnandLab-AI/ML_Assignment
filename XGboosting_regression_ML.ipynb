```python
import pandas as pd
dataset=pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)    ## coulumns name - sex_female	sex_male	smoker_no	smoker_yes
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
independent
```

```python
dependent=dataset[['charges']]
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
regressor=GradientBoostingRegressor(loss='uantile',criterion='friedman_mse',random_state=0,n_estimators=100)
regressor=regressor.fit(X_train,y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```
